```python
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def start(self):
        print(f"{self.make} {self.model} is starting...")

    def stop(self):
        print(f"{self.make} {self.model} is stopping...")

car1 = Car("Toyota", "Corolla")
car1.start()
car1.stop()
```

```python
class Animal:
    def make_sound(self):
        print("Some generic sound")

class Dog(Animal):
    def make_sound(self):
        print("Bark")

class Cat(Animal):
    def make_sound(self):
        print("Meow")

class Bird(Animal):
    def make_sound(self):
        print("Chirp")

animals = [Dog(), Cat(), Bird()]
for animal in animals:
    animal.make_sound()
```

```python
import numpy as np

arr = np.array([4, 12, 7, 9, 3])
print("Array:", arr)
print("Maximum value:", np.max(arr))
print("Minimum value:", np.min(arr))
print("Index of max:", np.argmax(arr))
print("Index of min:", np.argmin(arr))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = [1, 2, 3, 4, 5]
y = [5, 6, 2, 3, 13]
z = [2, 3, 3, 3, 5]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.subplot(1, 3, 1)
plt.scatter(x, y)
plt.title("Scatter")

plt.subplot(1, 3, 2)
plt.plot(x, y)
plt.title("Line")

plt.subplot(1, 3, 3)
plt.bar(x, y)
plt.title("Bar")

plt.tight_layout()
plt.show()
```
